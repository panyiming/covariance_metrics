```python
import pandas as pd
import numpy as np
```

```python
def read_csv(path, st='2000-01-01', et='2020-12-31'):
    code_price = {}
    code_shares = {}
    dates = []
    j = 0
    with open(path, 'rb') as f:
        for l in f:
            if j == 0:
                j += 1
                continue
            j +=1
            parts = l.decode().strip().split(',')
            date = parts[1]
            code = parts[-1]
            shares = int(parts[-2])
            adjust_p = float(parts[6])
            in_dates = (date>=st and date<=et)
            if code=='AAPL' and in_dates:
                dates.append(date)
       
            if in_dates and shares>1:
                if code not in code_price:
                    code_price[code] = []
                if code not in code_shares:
                    code_shares[code] = shares
                code_price[code].append(adjust_p)
    return code_price, code_shares, dates
```

```python
code_price, code_shares, dates = read_csv('./SP500_price.csv')
```

```python
j = 0
lens = []
for i in code_price:
    lens.append(len(code_price[i]))
    if len(code_price[i])>=5284:
        lens.append(len(code_price[i]))
        j +=1
        
print(j)
print(lens)
```

```python
data_table1 = np.zeros((5254,405))
num_lim = 3525
k = 0
names = []
for i in code_price:
    if len(code_price[i])>=num_lim:
        data_table1[:, k] = np.array(code_price[i])
        names.append(i)
        k +=1
df2 = pd.DataFrame(data=data_table1, index=dates, 
      columns=names)
df2.to_csv('SP405_20070103_20201231.csv')
```
